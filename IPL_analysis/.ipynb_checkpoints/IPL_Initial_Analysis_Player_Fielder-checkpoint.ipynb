```python
matches_batsman_run = ball_matches_df.groupby(['match_id','batting_team', 'batsman']).agg({'batsman_runs':'sum'}).reset_index()
```

```python
# matches_batsman_run
```

```python
ids = matches_batsman_run.groupby(['batting_team'])['batsman_runs'].transform(max) == matches_batsman_run['batsman_runs']
```

```python
matches_batsman_run_max =  matches_batsman_run[ids].sort_values(by='batsman_runs', ascending=False)
```

```python
matches_batsman_run_max
```

```python
matches_batsman_six = ball_matches_df[ball_matches_df.batsman_runs >= 6].groupby(['match_id','batting_team', 'batsman']).agg({'batsman_runs':'count'}).reset_index()
```

```python
matches_batsman_six
```

```python
ids = matches_batsman_six.groupby(['batting_team'])['batsman_runs'].transform(max) == matches_batsman_six['batsman_runs']
```

```python
matches_batsman_six_max= matches_batsman_six[ids].sort_values(by='batsman_runs', ascending=False)
```

```python
matches_batsman_six_max
```

```python
matches_batsman_run.groupby('batsman').agg({'batsman_runs':'mean'}).sort_values(by='batsman_runs', ascending=False).head()
```

```python
matches_batsman_six = ball_matches_df[ball_matches_df.batsman_runs >= 6].groupby(['match_id','batting_team', 'batsman']).agg({'batsman_runs':'count'}).reset_index()
```

```python
matches_batsman_run = ball_matches_df.groupby(['match_id','batting_team', 'batsman']).agg({'batsman_runs':'sum'}).reset_index()
```

```python
matches_batsman_run[matches_batsman_run.batsman_runs>=100].groupby('batsman').agg({'batsman_runs':'count'}).sort_values(by='batsman_runs',ascending=False)
```

```python
matches_batsman_run[matches_batsman_run.batsman_runs==0].groupby('batsman').agg({'batsman_runs':'count'}).sort_values(by='batsman_runs',ascending=False)
```

```python
matches_batsman_run[(matches_batsman_run.batsman_runs>=50) & (matches_batsman_run.batsman_runs<=100) ].groupby('batsman').agg({'batsman_runs':'count'}).sort_values(by='batsman_runs',ascending=False)
```
